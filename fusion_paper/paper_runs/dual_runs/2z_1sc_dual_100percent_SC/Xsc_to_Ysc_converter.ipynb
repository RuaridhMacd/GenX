```julia
using DataFrames
using CSV
```

```julia
home_dir = @__DIR__
var_data_dir = joinpath(home_dir, "Generators_variability_all.csv")
load_data_dir = joinpath(home_dir, "Load_data_all.csv")
```

```julia
# Load var_data_dir CSV file
var_data_all = CSV.read(var_data_dir, DataFrame) 
# Break up the data into an array of DataFrames, one for each year
T = 8760
var_data_byyear = [var_data_all[(idx-1)*T+1:idx*T,:] for idx in 1:20]
```

```julia
# Load load data
load_data_all = CSV.read(load_data_dir, DataFrame, header=true)
# Only keep the Load_MW_z2 column
load_data_all = load_data_all[:, :Load_MW_z2]
# Break up the data into an array of vectors, one for each year
T = 8760
load_data_byyear = [load_data_all[(idx-1)*T+1:idx*T] for idx in 1:20]
```

```julia
template_dir = joinpath(home_dir, "year_1")
```

```julia
for i = 2:20
    println("Creating year $i")
    # Copy the template dir into a new dir called "year_$i"
    run(`cp -r year_1 year_$i`)

    # Open the Generators_variability.csv file into temp_var
    # temp_var = CSV.read(joinpath(home_dir,"year_$i","Generators_variability.csv"), DataFrame)
    
    # Save the var_data_byyear[i] as the new Generators_variability.csv file
    CSV.write(joinpath(home_dir,"year_$i","Generators_variability.csv"), var_data_byyear[i], writeheader=true)

    # Open the Load_data.csv file into temp_load
    temp_load = CSV.read(joinpath(home_dir,"year_$i","Load_data.csv"), DataFrame)
    # Replace the zone 2 load with load_data_byyear[i]
    temp_load[:, :Load_MW_z2] = load_data_byyear[i]
    # Save the temp_load as the new Load_data.csv file
    CSV.write(joinpath(home_dir,"year_$i","Load_data.csv"), temp_load, writeheader=true)
end
```
